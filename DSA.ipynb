```python
class node:
  def __init__(self,data=None):
    self.data=data
    self.next=None

class linkedlist:
  def __init__(self):
    self.head=node()

  def insertback (self,data):
    if self.head==None:
      self.head=data
    new_node=node(data)
    temp=self.head
    while temp.next!=None:
      temp=temp.next
    temp.next=new_node

  def insertfront (self,data):
    if self.head==None:
      self.head=data
    new_node=node(data)
    temp=new_node
    temp.next=self.head
    self.head=new_node
  
  def insert(self , data , index):
    if self.head.data==None:
      self.head.data=data
    new_node = node(data)
    temp = self.head
    if index == 0:
      new_node.next = self.head
      self.head = new_node
    count=0
    while temp!=None:
      if count == index-1:
          new_node.next = temp.next
          temp.next = new_node
          break
      temp = temp.next
      count= count + 1
    else:
      print("sorry we cannot insert at this index because previous index doesnot exsist")

  def display(self):
    temp=self.head
    dis=[]
    while temp.next!=None:
      temp=temp.next 
      dis.append(temp.data)
    print(dis)

  def delete_end(self):
    temp=self.head
    while (temp.next.next):
      secondlast_node=temp.next
      temp=temp.next
    temp.next=None 
    print('node deleted from end') 
  
  def delete_beg(self):
    if self.head==None:
      print("no nodes exist")
    else:
      temp=self.head
      self.head=self.head.next
      temp=None
      print("node deleted from front")
  
  def delete_mid(self,index):
    if self.head==None:
      print("no node exist")
    count=0
    temp=self.head
    pre=temp
    while count<index:
      pre=temp
      temp=temp.next
      count+=1
    pre.next=temp.next
    temp.next=None
```
